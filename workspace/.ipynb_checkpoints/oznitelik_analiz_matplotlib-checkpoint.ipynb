```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df_path=r'C:\Users\tunahan.akyol\Desktop\lstm\saved_models\24w16s4elevelaccranangradrcs\20220209_1747\data'
df_batch_path=r'C:\Users\tunahan.akyol\Desktop\lstm\batch_test'
```

```python
df_train=pd.read_csv(df_path+"\\train.csv")
df_val=pd.read_csv(df_path+"\\val.csv")
df_test=pd.read_csv(df_batch_path+"\\extractedData_All_Test.csv")
df=pd.concat([df_train,df_val,df_test])
df_pos=df[df['label']==1]
df_neg=df[df['label']==0]
```

```python
def boxplot(feature,xlabel1,xlabel2):
    fig,ax=plt.subplots(1,2)
    sns.set(rc={'figure.figsize':(24,18)})
    sns.boxplot(x=df_pos[feature],ax=ax[0])
    sns.boxplot(x=df_neg[feature],ax=ax[1])
    ax[0].set(xlabel=xlabel1)
    ax[1].set(xlabel=xlabel2)
    sns.set_context("talk", font_scale=1.5, rc={"lines.linewidth":2})
    
```

```python
def histplot(feature,xlabel1,xlabel2,bins,df_pos,df_neg):
    fig,ax=plt.subplots(1,2)
    ax[0].hist(df_pos[feature],bins=bins)
    ax[1].hist(df_neg[feature],bins=bins)
    ax[0].set_xlabel(xlabel1)
    ax[1].set_xlabel(xlabel2)
    ax[0].set_ylabel("Number of Sample")
    
```

```python
def high_lowpass(feature,pos_high_value,pos_low_value,neg_high_value,neg_low_value,df_pos,df_neg):
    df_pos_high=len(df_pos[df_pos[feature]>pos_high_value].index)
    df_pos_low=len(df_pos[df_pos[feature]<pos_low_value].index)
    
    df_neg_high=len(df_neg[df_neg[feature]>neg_high_value].index)
    df_neg_low=len(df_neg[df_neg[feature]<neg_low_value].index)
    
    len_pos=len(df_pos.index)
    len_neg=len(df_neg.index)
    print(f"total_pos_number: {len(df_pos.index)}, number of {feature} sample bigger than {pos_high_value}: {df_pos_high}")
    print(f"total_pos_number: {len(df_pos.index)}, number of {feature} sample smaller than {pos_low_value}: {df_pos_low}")
    
    print(f"total_neg_number: {len(df_neg.index)}, number of {feature} sample bigger than {neg_high_value}: {df_neg_high}")
    print(f"total_neg_number: {len(df_neg.index)}, number of {feature} sample smaller than {neg_low_value}: {df_neg_low}")
    
    df=pd.DataFrame(data={'thresholds':[pos_high_value,pos_low_value,neg_high_value,neg_low_value],
                          'total_sample':[len_pos,len_pos,len_neg,len_neg],
                          'filtered_sample':[df_pos_high,df_pos_low,df_neg_high,df_neg_low],
                         '%percentage':[(df_pos_high/len_pos)*100,(df_pos_low/len_pos)*100,(df_neg_high/len_neg)*100,(df_neg_low/len_neg)*100]},
                    
                    index=['pos_higher_threshold','pos_lower_threshold','neg_higher_threshold','neg_lower_threshold'],
                    columns=['thresholds','total_sample','filtered_sample','%percentage'])
    return df
```

```python
df_rcs_filtresi=high_lowpass('rcs',0.25,0.25,0.25,0.25,df_pos,df_neg)
df_rcs_filtresi
```

```python
rcs_indexes_pos=df_pos[df_pos['rcs']>0.25]
rcs_indexes_neg=df_neg[df_neg['rcs']>0.25]
histplot('range','drone range','negative range',10,rcs_indexes_pos,rcs_indexes_neg)
```

##  RCS'i 0.25 den büyük ve menzili 500 den küçük sample sayısı

```python
rcs_range_filtered_pos=df_pos[(df_pos['rcs']>0.25) & (df_pos['range']<500)]
rcs_range_filtered_neg=df_neg[(df_neg['rcs']>0.25) & (df_neg['range']<500)]
df_rcs_range_filtered=pd.DataFrame(data={'threshold':[0.25,0.25],'total_sample':[len(df_pos.index),len(df_neg.index)],'filtered_sample':[len(rcs_range_filtered_pos.index),len(rcs_range_filtered_neg.index)]},
                                  index=['pos','neg'])
df_rcs_range_filtered
```

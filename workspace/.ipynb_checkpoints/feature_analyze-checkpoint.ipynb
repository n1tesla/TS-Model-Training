## RCS analysis

```python
fig,ax=plt.subplots(1,2)
sns.boxplot(x=df_pos['rcs'],ax=ax[0])
sns.boxplot(x=df_neg['rcs'],ax=ax[1])
ax[0].set(xlabel='POSITVE RCS')
ax[1].set(xlabel='NEGATIVE RCS')
sns.set_context("talk", font_scale=1.5, rc={"lines.linewidth":2})
# sns.set(rc={'figure.figsize':(24,18)})

#plt.figure(figsize = (12,8))
```

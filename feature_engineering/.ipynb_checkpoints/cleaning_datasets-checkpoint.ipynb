# Delete unnecessary data

if u want to use just positive data or negative data in dataframe choose label. Ex label=1 you get all positive data and delete all negative data.

```python
def get_all_pos_or_neg(df,label):
    df=df[df['label']==label]
    return df
```

specify the rectangular area (x,y positions) to delete unwanted data. It'll remove tracks inside the (x1,x2,y1,y2) positions

```python
def del_inside_rect(df,x1,x2,y1,y2):
    df.drop(df.index[(df['X_labeled']>x1) & (df['X_labeled']<x2) & (df['Y_labeled']>y1) & (df['Y_labeled']<y2)],inplace=True)
    print(f"the data in x:({x1}-{x2}),y:({y1}-{y2}) positions deleted") 
    return df                                                                                   
```

specify the id_no to delete irradiated data. Deletes all id

```python
def del_based_on_id(df,id_no):
    print(f"ID_no: {id_no}, len: {len(df.index[df['ID_labeled']==id_no])} sample deleted via based on id")
    df.drop(df.index[df['ID_labeled']==id_no],inplace=True)
    return df
```

specify the x,y positions to delete clustered data and if it has the same id outside of the specified positions, also deletes it.

```python
def del_ids_in_rect(df,x1,x2,y1,y2):
    del_index=df[(df['X_labeled']>x1) & (df['X_labeled']<x2) & (df['Y_labeled']>y1) & (df['Y_labeled']<y2)]['ID_labeled'].unique()
    will_drop_df=df['ID_labeled'].isin(del_index)
    print(f"{len(df.index[will_drop_df])} sample deleted via based_on_id_and_pos")
    df.drop(df.index[will_drop_df],inplace=True)
    return df
```

specify the id number and x,y pos to delete bad data. It deletes bad part of the id

```python
def del_partial_id(df,id_no,x1,x2,y1,y2):
    del_index=df[(df['X_labeled']>x1) & (df['X_labeled']<x2) & (df['Y_labeled']>y1) & (df['Y_labeled']<y2) & (df['ID_labeled']==id_no)]['ID_labeled'].index
    print(f"{len(del_index)} samples  deleted via del partial id")
    df.drop(del_index,inplace=True)
    return df
```

clear the data outside of the radar range, To delete too far tracks.

```python
def del_based_on_xy(df,x1,x2,y1,y2):
    del_index1=df[df['X_labeled']<x1].index
    df.drop(del_index1,inplace=True)
    del_index2=df[df['X_labeled']>x2].index
    df.drop(del_index2,inplace=True)
    print(f"{len(del_index1)+len(del_index2)} sample deleted via_based_on_x")
    
    del_index3=df[df['Y_labeled']<y1].index
    df.drop(del_index3,inplace=True)
    del_index4=df[df['Y_labeled']>y2].index
    df.drop(del_index4,inplace=True)
    print(f"{len(del_index3)+len(del_index4)} sample deleted via_based_on_y")

    return df
```

remove the negative data whose data jumped

```python
def del_jumped_data(df):
    df_neg=df[df['label']==0]
    df_pos=df[df['label']==1]
    jumped_data=df_neg['jumped/ok'].isin(['jumped'])
    unique_ids=df_neg.loc[df_neg.index[jumped_data]]['ID_labeled'].unique()
    print(f"{len(df_neg.index[df_neg['ID_labeled'].isin(unique_ids)])} negative data deleted via jumped_data")
    df_neg.drop(df_neg.index[df_neg['ID_labeled'].isin(unique_ids)],inplace=True)
    df=pd.concat([df_pos,df_neg])
    return df
```

delete data if ID has less than 12 samples. It is hard to augment has less than 12 samples data

```python
def del_less_than_12_samples(df):
    df_less=df['ID_labeled'].value_counts()<12
    ids=df_less[df_less].index
    print(f"{len(df.index[df['ID_labeled'].isin(ids)])} sample deleted via less_than_12 samples")
    df.drop(df.index[df['ID_labeled'].isin(ids)],inplace=True)
    return df
```

```python
def create_positions_subset(ID):
    df_subset=df[df['ID_labeled']==ID]
    pos_x=df_subset['X_labeled']
    pos_y=df_subset['Y_labeled']
    return pos_x,pos_y
```

```python
def histplot(feature,xlabel1,xlabel2,df_pos,df_neg):
    fig,ax=plt.subplots(1,2)
    ax[0].hist(df_pos[feature])
    ax[1].hist(df_neg[feature])
    ax[0].set_xlabel(xlabel1)
    ax[1].set_xlabel(xlabel2)
    ax[0].set_ylabel("Number of Sample")
```

```python
def boxplot(feature,xlabel1,xlabel2,df_pos,df_neg):
    fig,ax=plt.subplots(1,2)
    sns.set(rc={'figure.figsize':(24,18)})
    sns.boxplot(x=df_pos[feature],ax=ax[0])
    sns.boxplot(x=df_neg[feature],ax=ax[1])
    ax[0].set(xlabel=xlabel1)
    ax[1].set(xlabel=xlabel2)
    sns.set_context("talk", font_scale=1.5, rc={"lines.linewidth":2})
```

```python
def plot_posxy(df):
    df_pos=df[df['label']==1]
    df_neg=df[df['label']==0]
    pos_x_1=df_pos['X_labeled']
    pos_y_1=df_pos['Y_labeled']
    pos_x_0=df_neg['X_labeled']
    pos_y_0=df_neg['Y_labeled']
    plt.rcParams['figure.figsize']=(20,10)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.rcParams.update({'font.size': 22})
    plt.title('Radar Positive and Negative Class XY')
    plt.scatter(pos_x_1,pos_y_1,marker='.',linewidths=1,c='r',label='drone')
    plt.scatter(pos_x_0,pos_y_0,marker='.',linewidths=1,c='b',label='neg')
    plt.legend()
    plt.tight_layout()
    plt.savefig("filtered_data_posxy")
```

```python
def save_csv(df):
    print(f"length of positive data: {len(df[df['label']==1].index)}, length of negative data: {len(df[df['label']==0].index)}")
    df.to_csv("filtered_data.csv")
```
